```python
# Regular expression library
import re

sum_email = 0
Split_email = []

for email in Ticket["From Email"]:
    email = email.split('@')
    Split_email.append(email[1])
    sum_email += 1
```

```python
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date
```

```python
Split_email = pd.DataFrame(Split_email)
```

```python
Split_email['Email @domains'] = Split_email
```

```python
del Split_email[0]
```

```python
Ticket = Ticket.assign(Email_Domains = Split_email)
```
