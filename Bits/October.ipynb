```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Oct = pd.read_csv("October.csv")
```

```python
Oct.info()
```

```python
# Regular expression library
import re

sum_email = 0
Split_email = []

for email in Oct["From Email"]:
    email = email.split('@')
    Split_email.append(email[1])
    sum_email += 1
```

```python
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date
```

```python
Split_email = pd.DataFrame(Split_email)
```

```python
Split_email['Email @domains'] = Split_email
```

```python
Split_email
```

```python
del Split_email[0]
```

```python
Oct=Oct.assign(Email_Domains = Split_email)
```

```python
Oct.head(2)
```

```python
October = Oct[Oct["Date Created"] >= '2020/10/01 00:00']
```

```python
October
```

```python
October['Closed Date']= pd.to_datetime(October['Closed Date']).dt.date 
```

```python
October['Date Created']= pd.to_datetime(October['Date Created']).dt.date 
```

```python
October
```

```python
October['diff_days'] = abs(October['Closed Date'] - October['Date Created'])
```

```python
October.head(2)
```

```python
October.to_csv('October_Cleaned_Dataset.csv')
```
